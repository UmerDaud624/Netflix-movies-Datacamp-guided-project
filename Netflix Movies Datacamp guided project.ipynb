```python
years = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
```

```python
durations = [103, 101, 99, 100, 100, 95, 95, 96, 93, 90]
```

```python
movie_dict = {
    'years' : years,
     'durations' : durations
}
```

```python
print(years)
```

```python
print(durations)
```

```python
print(movie_dict)
```

```python
import pandas as pd
```

```python
durations_df = pd.DataFrame(movie_dict)
```

```python
print(durations_df)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title('Netflix Movie Durations 2011-2020')
plt.plot(durations_df.years, durations_df.durations);
```

### The duration of movies have been reduced over the period, however there was a little increase around 2013 and 2017 but then again started to reduce in 2015 and 2018 respectively.

```python
netflix_df = pd.read_csv('netflix_data.csv')
```

```python
netflix_df.head(5)
```

```python
netflix_df_movies_only = netflix_df[netflix_df.type == 'Movie']
```

```python
netflix_df_movies_only
```

```python
netflix_movies_col_subset = netflix_df_movies_only.copy()
netflix_movies_col_subset.drop(columns=['show_id', 'type', 'director', 'cast', 'date_added', 'description'], inplace=True)
```

### Copy function is used to make a shallow copy and the columns are dropped which are not needed.

```python
netflix_movies_col_subset.head(5)
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(9, 6))
plt.title('Movie Duration by Year of release')
sns.scatterplot(x='release_year', y='duration', data= netflix_movies_col_subset);
plt.grid('darkgrid')
```

```python
netflix_movies_col_subset.info()
```

```python
short_movies = netflix_movies_col_subset[netflix_movies_col_subset.duration < 60]
```

```python
short_movies.head(20)
```

```python
colors = []
for i in netflix_movies_col_subset.genre:
  if i == 'Children':
    colors.append('red')
  elif i == 'Documentaries':
    colors.append('blue')
  elif i == 'Stand-Up':
    colors.append('green')
  else:
    colors.append('black')
```

```python
for i in range(0, 10):
  print(colors[i])
```

```python
plt.figure(figsize=(9, 6))
plt.title('Movie duration by Year of release')
plt.scatter(x= netflix_movies_col_subset.release_year, y= netflix_movies_col_subset.duration, c= colors);
plt.grid('darkgrid')
```

### There were some short lenght movies around 1945 but then the duration of movies increased around 1960. However, after 2000 there are many short length movies around 60-90 minutes made on netflix and they are mainly stand-ups, Documentaries and for Children. Moreover, there are many long lengths movies on netflix which are represented in black in graph and they are of genre, Drama, action , horror, comedy etc. We can conclude that netflix is gradually producing less duration movies but still it does not have stopped producing long duration movies, they are also in bulk.
